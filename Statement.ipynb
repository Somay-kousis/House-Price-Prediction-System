## Project: House Price Prediction with Linear & Ridge Regression

### Dataset: 
**Boston Housing Dataset** (built into sklearn) or **Ames Housing Dataset** from Kaggle
- **Why**: Perfect size for learning, mix of numerical/categorical features, well-documented

### Tasks to Accomplish (in order):

1. **Data Loading & Exploration**
   - Load the dataset using pandas
   - Explore data shape, info, and basic statistics
   - Create visualizations showing price distributions and feature relationships

2. **Data Preprocessing**
   - Handle missing values (if any)
   - Create correlation matrix heatmap
   - Identify the most correlated features with price

3. **Feature Engineering**
   - Select 3-5 most important features for your model
   - Scale features using StandardScaler
   - Split data into train/test sets (80/20)

4. **Model Implementation**
   - Implement Linear Regression from scratch using numpy (just the math)
   - Implement the same using sklearn.linear_model.LinearRegression
   - Compare both results

5. **Ridge Regression Analysis**
   - Implement Ridge regression with different alpha values [0.1, 1, 10, 100]
   - Plot how coefficients change with different alpha values
   - Use cross-validation to find the best alpha

6. **Model Evaluation & Comparison**
   - Calculate MSE, RMSE, and R² for both models
   - Create residual plots to check model assumptions
   - Visualize actual vs predicted prices

7. **Insights & Documentation**
   - Write conclusions about which model performs better
   - Explain the effect of regularization on your coefficients
   - Create a simple function that can predict house prices given input features

### Expected Outcome:
By the end, you'll have a complete regression project that demonstrates understanding of both linear and ridge regression, setting you up perfectly for Week 3's classification focus.

This should take about 4-6 hours spread over 2 days. Focus on understanding WHY each step matters rather than just getting it to work!
